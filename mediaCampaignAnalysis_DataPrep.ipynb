📌 Overview: Data Cleaning & Preprocessing
Plan of action details:

✅ Step 1: Assure all columns are numeric for analysis

✅ Step 2: Handle Missing Values (Income)
Since only 24 rows have missing Income, we can:

Fill them with the median income (preferred to avoid skewing data).
Drop not useful.

✅ Step 3: Convert Date Column (Dt_Customer)
convert Dt_Customer to datetime format.
Create Customer_for_Days field.

Why?

Customer_for_Days tells how long the customer has been with the company in days since they joined/purchased.

✅ Step 4: Remove Outliers (Year_Birth)
remove unrealistic ages (e.g., before 1920 or after 2000).

✅ Step 5: Create New Features
1.Total Amount Spent by Customer
2.Total Purchases Across Channels
3.Engagement Rate (Website Visits / Total Purchases)

✅ Step 6: Encode Categorical Variables
Convert Education and Marital_Status into numerical values.

📌Final Step: Save the Cleaned Data
store this cleaned dataset in BigQuery for future use.

```python
# Convert relevant columns to numeric
columns_to_sum = ["MntWines","MntFruits","MntMeatProducts","MntFishProducts","MntSweetProducts","MntGoldProds","NumDealsPurchases","NumWebPurchases","NumCatalogPurchases","NumStorePurchases","NumWebVisitsMonth"]

for col in columns_to_sum:
    df[col] = pd.to_numeric(df[col], errors="coerce")  # Convert to numeric, set invalid values as NaN

# Now perform the summation correctly
df["Total_Amount_Spent"] = df[columns_to_sum].sum(axis=1)

# verify results
print(df[["Total_Amount_Spent"]].head())  # Check if values are correct
```

```python
print(df.dtypes)  # Check data types of all columns
print(df.head())  # Preview the first few rows
```

```python
#Fill missing income values with median values
df.fillna({"Income": df["Income"].median()}, inplace=True)
#Verify
print("missing values handled: ", df.isnull().sum())
```

```python
# Convert Dt_Customer to datatime
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"])

#Extract useful data-based feature
df["Customer_for_Days"] = (pd.to_datetime("today") - df["Dt_Customer"]).dt.days

#Verify
print("Date column converted")
df[["Dt_Customer", "Customer_for_Days"]].head()
```

```python
#Define valid age range
df = df[(df["Year_Birth"] >= 1928) & (df["Year_Birth"] <= 2000)]

#verify
print(f"Outliers removed. Reamianing records: {len(df)}")
```

```python
#Create new features for future analytical use

##Total amount spent by customer
df["Total_Spending"] = df[
    ["MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts", "MntSweetProducts", "MntGoldProds"]
].sum(axis=1)

#verify
print(df[["Total_Spending"]].head())

```

```python
##Total Purchases Across Channels
df["Total_Purchases"] = df[
    ["NumDealsPurchases", "NumWebPurchases", "NumCatalogPurchases", "NumStorePurchases"]
].sum(axis=1)

#verify
df["Total_Purchases"].head()
```

```python
##Engagement Rate(Website visits/Total Purchases)
df["Engagement_Rate"] = df["NumWebVisitsMonth"] / (df["Total_Purchases"] + 1)

#verify
df["Engagement_Rate"].head()
```

```python
#Encode Categorical Variables Education and Marital_Status into numerial values for ML analysis
df = pd.get_dummies(df, columns=['Education_Basic','Education_Graduation', 'Education_Master', 'Education_PhD','Marital_Status_Alone', 'Marital_Status_Divorced','Marital_Status_Married', 'Marital_Status_Single','Marital_Status_Together', 'Marital_Status_Widow','Marital_Status_YOLO'], drop_first=True)
```

```python
# Rename columns to remove "_True"
df.columns = df.columns.str.replace("_True", "")

# Convert boolean values to integers (0/1)
df = df.astype(int)

#verify
print("✅ Column names cleaned and encoded variables converted to 0/1!")
df.head()
```
